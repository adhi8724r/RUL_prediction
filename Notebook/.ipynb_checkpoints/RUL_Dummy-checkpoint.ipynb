```python
# Expand dictionary values into separate columns
df_charge_expanded = df_charge.join(pd.json_normalize(df_charge['data']))
df_discharge_expanded = df_discharge.join(pd.json_normalize(df_discharge['data']))

# Drop the original 'data' column as it's no longer needed
df_charge_expanded.drop(columns=['data'], inplace=True)
df_discharge_expanded.drop(columns=['data'], inplace=True)

# Display the first few rows
df_charge_expanded.head(), df_discharge_expanded.head()
```

```python
from scipy.stats import f_oneway
import numpy as np

# Extract relevant features for charge data
df_charge_extracted = df_charge_expanded.copy()
df_charge_extracted["Voltage_measured"] = df_charge_extracted["Voltage_measured"].apply(np.mean)
df_charge_extracted["Current_measured"] = df_charge_extracted["Current_measured"].apply(np.mean)
df_charge_extracted["Temperature_measured"] = df_charge_extracted["Temperature_measured"].apply(np.mean)
df_charge_extracted["Current_charge"] = df_charge_extracted["Current_charge"].apply(np.mean)
df_charge_extracted["Voltage_charge"] = df_charge_extracted["Voltage_charge"].apply(np.mean)

# Extract relevant features for discharge data
df_discharge_extracted = df_discharge_expanded.copy()
df_discharge_extracted["Voltage_measured"] = df_discharge_extracted["Voltage_measured"].apply(np.mean)
df_discharge_extracted["Current_measured"] = df_discharge_extracted["Current_measured"].apply(np.mean)
df_discharge_extracted["Temperature_measured"] = df_discharge_extracted["Temperature_measured"].apply(np.mean)
df_discharge_extracted["Current_load"] = df_discharge_extracted["Current_load"].apply(np.mean)
df_discharge_extracted["Voltage_load"] = df_discharge_extracted["Voltage_load"].apply(np.mean)

# The target variable for RUL prediction is 'Capacity' (only in discharge data)
df_discharge_extracted["Capacity"] = df_discharge_extracted["Capacity"]
```

```python
df_discharge_extracted
```

```python


# Perform One-Way ANOVA test to determine feature importance
anova_results = {}
features = ["Voltage_measured", "Current_measured", "Temperature_measured", "Current_load", "Voltage_load"]

for feature in features:
    f_stat, p_value = f_oneway(df_discharge_extracted[feature], df_discharge_extracted["Capacity"])
    anova_results[feature] = (f_stat, p_value)

# Display ANOVA results
print(anova_results)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# Select important features for training
features = ["Voltage_measured", "Current_measured", "Temperature_measured", "Voltage_load", "Current_load"]
target = "Capacity"

# Split data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(df_discharge_extracted[features], df_discharge_extracted[target], test_size=0.2, random_state=42)

# **1️⃣ Train Random Forest Model**
rf_model = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
rf_model.fit(X_train, y_train)
rf_preds = rf_model.predict(X_test)

# **2️⃣ Train Support Vector Machine (SVM) Model**
scaler = StandardScaler()  # Normalize features for SVM
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm_model = SVR(kernel="rbf", C=10)
svm_model.fit(X_train_scaled, y_train)
svm_preds = svm_model.predict(X_test_scaled)

# **Evaluate Models**
rf_mse = mean_squared_error(y_test, rf_preds)
rf_r2 = r2_score(y_test, rf_preds)

svm_mse = mean_squared_error(y_test, svm_preds)
svm_r2 = r2_score(y_test, svm_preds)

# Display results
print("Random Forest Results: MSE =", rf_mse, ", R² =", rf_r2)
print("SVM Results: MSE =", svm_mse, ", R² =", svm_r2)
```
